```python
print("DIAMOND RUSH")
import time
import random

Player_inventory = []
progress = []
game_over = False

highlighted_text = "\033[1m"
reset_formatting = "\033[0m"

def save_progress_to_file(filename, progress_list):
    with open(filename, 'w') as file:
        for step in progress_list:
            file.write(step + '\n')
    print(f"Progress saved to '{filename}'.")


def load_progress_from_file(filename):
    progress = []
    try:
        with open(filename, 'r') as file:
            for line in file:
                progress.append(line.strip())
        print(f"Progress loaded from '{filename}'.")
        return progress
    except FileNotFoundError:
        print(f"'{filename}' not found. Starting a new game.")
        return []


def story_mode(text, highlight=False, font_size=None):
    if highlight:
        text = highlighted_text + text + reset_formatting
    if font_size:
        text = f"\033[{font_size}m{text}{reset_formatting}"
    for char in text:
        print(char, end='', flush=True)
        time.sleep(0.04)
    print()
    progress.append(text)

def interactive_choice(options, prompt):
    while True:
        for i, option in enumerate(options, start=1):
            print(f"{i}. {option}")
        try:
            choice = int(input(prompt))
            if 1 <= choice <= len(options):
                return choice
            else:
                print("Invalid choice. Please select a valid option.")
        except ValueError:
            print("Invalid input. Please enter a number.")

def add_to_inventory(item):
    Player_inventory.append(item)
    print(f"You added {item} to your inventory.")

def remove_from_inventory(item):
    if item in Player_inventory:
        Player_inventory.remove(item)
        print(f"You used {item}.")
    else:
        print(f"You don't have {item} in your inventory.")

def show_inventory():
    print("Your inventory:")
    for item in Player_inventory:
        print(f"- {item}")
        
        
def display_progress():
            print("Your progress so far:")
            for i, step in enumerate(progress, start=1):
                print(f"{i}. {step}")

story_mode("This game adapts to the choices you make, and the story is tailored by how you play!!", highlight=True, font_size="31")
story_mode("Your name is Cal. You are an international thief who works as a professional undercover criminal in the world's most powerful mafia.")
story_mode("When you are off duty, you work as a normal employee in an IT company as a sales manager.")
story_mode("Suddenly,one morning, you get a call from the CIA, and you are told to urgently meet them. When asked for a reason, they deny to mention it and call you to their main office.")
story_mode("You remember that you had recently been on a mission for stealing the world's costliest diamond and deliver it to the mafia.")
story_mode("Realizing that you had been on a deadly mission, you cannot face the CIA, and you plan for an escape to a different country as a disguise.")
story_mode("Making an attempt to escape will be dangerous as it will involve one of the best officers of the CIA, including the undercovers.")
story_mode("To be able to escape, you will need strong support and a good inventory.")
Player_inventory = ['Diamond_bag', 'P92']
show_inventory()

while True:
    story_mode("The game starts now. You have a choice to make:")
    choice = interactive_choice([("Meet the CIA", "1"), ("Don't meet the CIA", "2")], "Enter the choice to make: ")

    if choice == 1:
        story_mode("You will be heading to the CIA office to meet the head officer.")
        story_mode("You call back the CIA head and tell them that you are on your way to meet them.")
        story_mode("The CIA send 2 of their officers to pick you up, and you are ready to head to their office.")
        story_mode("But before you leave, you need to check that you have all the important documents for your identity verification.")
        check_documents = ['Employee ID Card', 'Passport', 'Nationality Card', 'Bank Details']
        story_mode("You have the following important documents to verify that you are a common working man.")
        print(check_documents)
        story_mode("You reach the office and enter the head officer's room.")
        story_mode("He makes you sit down on an interrogation chair and starts asking you questions related to a diamond theft case.")
        story_mode("Hearing that, you start sweating and complain that you have nothing to do with it.")
        story_mode("You also demand them to check your documents in order to verify your identity.")
        choice = interactive_choice([("Hand over the documents", "1"), ("Give your statement and surrender", "2")], "Enter your choice: ")

        if choice == 1:
            story_mode("You hand over your documents for verification, and the officer checks if all documents are verified by the government and stay true to your identity.")
        else:
            story_mode("You surrender to the CIA in fear that you would be sentenced to life imprisonment, and you make up a story and present it as your statement so that your punishment will be far less than what you would get for not surrendering.")
            story_mode("You get a punishment of 10 years in jail for international theft offense and a fine of $100,000!")
            story_mode("Game Over!")
            break

        story_mode("The CIA officer sends you back, and without you knowing, he also sends an undercover agent as your driver to keep an eye on you.")
        story_mode("Now you call the mafia head in charge of this diamond theft case and tell him that you are about to make an escape.")
        story_mode("But what you do not know is that the undercover agent had secretly put a mic in your shirt which is connected to his hearing piece.")
        story_mode("Hearing all this, the agent decides to follow you and tracks you down to a godown which has been abandoned for years now.")
        story_mode("This is where the mafia meetings and secret missions are assigned.")
        story_mode("You and your head meet and plan on an escape outside the country, to France.")
        story_mode("The undercover agent takes your pictures and records your audios and sends it to the CIA Head Officer Mr. Johnson.")
        story_mode("The mafia head Mr. Terry decides to send you off in a helicopter which is privately owned and licensed by him.")
        story_mode("Mr. Terry hands you a special dagger and a bulletproof vest and a chemical which is helpful to put anyone to sleep, just in case if you need it.")
        story_mode("You can select what you want...")

        while True:
            choice = interactive_choice([("Take Special Dagger and Chemical Potion", "1"), ("Take Bulletproof Vest", "2"), ("Take Nothing", "3"), ("Done", "4")], "What do you want? ")

            if choice == 1:
                add_to_inventory("Special Dagger")
                add_to_inventory("Chemical Potion")
            elif choice == 2:
                add_to_inventory("Bulletproof Vest")
            elif choice == 3:
                print("You chose nothing. Can't Continue. Re-select!")
            else:
                print("Exiting inventory selection...")
                break

        show_inventory()

        story_mode("The same evening you are off in the helicopter.")
        story_mode("What you do not know is the undercover agent is the pilot.")
        story_mode("You think that it's an easy escape, but the undercover had informed the CIA to be present near the France border so that you can no longer escape and will soon be in the hands of the CIA.")
        story_mode("Mid-air, when you notice a tag on the pilot's right hand that he did not cover and you see a symbol which resembles the CIA logo.")
        story_mode("You come to know that he is an undercover agent that the CIA had sent.")
        story_mode("You don't panic and keep your calm and come up with 2 plans...")

        choice = interactive_choice([("Execute Plan 1", "1"), ("Execute Plan 2", "2")], "Which plan do you want to execute? ")

        if choice == 1:
            story_mode("You take your dagger out and stab him on the internal jugular vein which causes the most fatal damage to a person.")
            story_mode("Now the helicopter is in your control, and you can make an escape.")
            story_mode("You acquire the information that the undercover agent had been directing the CIA Head Officer.")
            story_mode("You come to know that the CIA team was informed well in advance that you were planning to enroute and escape to France.")
            story_mode("So, you change your plans and contact the mafia head that you will be heading to Spain.")
            story_mode("Once you cross the borders you successfully managed to dodge the CIA!")
            story_mode("")
            
        else:
            story_mode("You decide to make the pilot unconscious and take possession of the helicopter.")
            story_mode("But the pilot already knew about the chemical and had made preparations in advance.")
            story_mode("It's game over for you.")
            story_mode("You are handed over to the CIA team and sentenced to life imprisonment for fraudulent behavior and international theft offense.")

    elif choice == 2:
        story_mode("You decided not to meet the CIA and instead contact your mafia and make preparations for your escape to France.")
        story_mode("The CIA track your phone details and come to your house for an arrest.")
        story_mode("When you get to know the CIA have arrived at your house, you plan to escape from the balcony and take your car away from them.")
        story_mode("Hearing the noise of Your car, the CIA track the number down and chase you from all sides.")
        story_mode("At last, near an attic when there is no more road left, you have been surrounded from all sides. It's Game Over!")
        story_mode("After a few months, you get a chance to escape from jail.")
        story_mode("You take your chances and manage to escape, leaving behind your old identity.")
        story_mode("With your underground connections, you establish a new life in a different country, living under a new identity.")
        story_mode("The past is behind you, and you decide to lead an honest life, far from the criminal world.")

    story_mode("Congratulations, you have reached the end of your adventure!")

      
    
    restart = input("Do you want to play again? (yes/no): ")
    if restart.lower() != 'yes':
        break
display_progress()
```
